```python
import matplotlib.pyplot as plt
import numpy as np

def create_title_image(title, max_width=1080, font_size=25, rotation=0):
    plt.rcParams['font.sans-serif'] = ['Segoe UI Emoji']
    plt.rcParams['axes.unicode_minus'] = False
    fig, ax = plt.subplots(figsize=(max_width / 100, 0.5))
    ax.text(0.5, 0.5, title, fontsize=font_size, ha='center', va='center', rotation=rotation, color='black', alpha=0.8)
    ax.axis('off')
    fig.canvas.draw()
    title_image = np.frombuffer(fig.canvas.tostring_rgb(), dtype=np.uint8)
    title_image = title_image.reshape(fig.canvas.get_width_height()[::-1] + (3,))
    plt.show(fig)
    return title_image

title = "PLANLOS DURCH DIE GEGEND WETZEN üò¨üèÉüèª‚Äç‚ôÇÔ∏èüòá"
title_image = create_title_image(title, rotation=0) 

```
