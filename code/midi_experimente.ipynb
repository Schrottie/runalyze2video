```python
import mido

# MIDI-Datei laden
mid = mido.MidiFile('flohwalzer_downloaded.mid')

# Liste zum Speichern von Noten und deren Dauer
notes_and_durations = []

# Durch die Tracks der MIDI-Datei iterieren
for track in mid.tracks:
    # Durch die Nachrichten des Tracks iterieren
    current_time = 0
    for msg in track:
        # Wenn es sich um eine Notenon-Nachricht handelt
        if msg.type == 'note_on':
            # Die Note und deren Dauer extrahieren und zur Liste hinzufügen
            notes_and_durations.append((msg.note, msg.velocity, msg.time))
            current_time = 0
        else:
            # Die Zeit der Nachricht zur aktuellen Zeit hinzufügen
            current_time += msg.time

# Ausgabe der extrahierten Noten und deren Dauer
print(notes_and_durations)
```
